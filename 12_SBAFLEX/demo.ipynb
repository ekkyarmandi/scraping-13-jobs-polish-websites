```python
import requests
from bs4 import BeautifulSoup
import json, re, os

headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.82 Safari/537.36"}

def render_html(url):
    req = requests.get(url,headers=headers)
    return BeautifulSoup(req.text,"html.parser")

def get_details(page):
    detail_dict = {}
    ul = page.find("ul",class_="job-details")
    for li in ul:
        text = li.get_text().strip()
        if ":" in text:
            key = text.split(":")[0].strip()
            value = text.split(":")[1].strip()
            detail_dict.update({key:value})
    for li in ul.find_next("ul"):
        text = li.get_text().strip()
        if ":" in text:
            key = text.split(":")[0].strip()
            value = text.split(":")[1].strip()
            detail_dict.update({key:value})
    return detail_dict
```

```python
urls = []

print("Page:",end=" ")

i = 1
max_target = 500
while len(urls) < max_target:
    print(i,end=",")
    url = f"https://holandia.jober.pl/?p={i}"
    page = render_html(url)
    a_tags = page.find("table",class_="job-results").find_all("a")
    urls.extend(a_tags)
    i += 1
    
jobs = []
for i in range(len(urls)):
    url = urls[i]['href']
    title = urls[i]['title']
    e = {"url": url, "title": title}
    jobs.append(e)
    
print("\b. done.")
```

```python
data = []

print("Index:",end=" ")

i = 1
for e in jobs[:500]:
    
    print(i,end=",")
    
    url = e['url']
    title = e['title']
    try:
        page = render_html(url)
        detail_dict = get_details(page)
    except: detail_dict = None
    
    if detail_dict != None:
    
        try: agency = detail_dict['Firma']
        except: agency = None
        try: wage = detail_dict['Pensja']
        except: wage = None
        try: city = detail_dict['Miejscowość']
        except: city = None
        try: country = detail_dict['Państwo']
        except: country = None

        information = {
            "Link": url,
            "Job Title": title,
            "Agency (Firma)": agency,
            "Wage (Pansja)": wage,
            "City (Miejscowosc)": city,
            "Country (Panstwo)": country
        }

        data.append(information)
    
    i += 1

print("\b.")
```

```python
import pandas as pd

df = pd.DataFrame(data)
# df.to_excel("example.xlsx",sheet=2,index=False)
```
